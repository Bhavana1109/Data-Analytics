### Question Paper Analysis

### Introduction:
The dataset represents the performance or score data of multiple individuals across various questions or modules in a test or survey. 
Each row corresponds to a unique individual, and the columns represent the scores for different questions or modules. 
The "Total" column indicates the total score obtained by the individual, while the remaining columns (Q1aM4, Q1bM6, Q2aM6, etc.) 
represent scores for specific questions or modules, with the labels suggesting the question number and possible marks assigned 
(e.g., Q1aM4 might represent Question 1 part (a) with a maximum of 4 marks). Some values are missing (represented as NaN), 
indicating that the participant may have skipped the question or the data was not recorded.

```python
 df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
 df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
 plt.figure(figsize=(8, 6))
 plt.scatter(df['Q1'], df['Q2'], color='purple')
 plt.title('Scatter Plot of Marks: Q1 vs Q2')
 plt.xlabel('Marks in Q1')
 plt.ylabel('Marks in Q2')
 plt.grid(True)
 plt.tight_layout()
 plt.show()
```

```python
The scatter plot generated between the marks of Q1 and Q2 provides a visual representation
of the relationship between the two questions. Each point in the plot represents the marks
obtained by a student in both Q1 and Q2. The x-axis shows the marks for Q1, while the y
axis represents the marks for Q2.
```

```python
df['Q3'] = df['Q3aM5'] + df['Q3bM5']
mean_Q3 = df['Q3'].mean()
median_Q3 = df['Q3'].median()
plt.figure(figsize=(8, 6))
plt.hist(df['Q3'], color='grey', bins=10, edgecolor='black', alpha=0.8)
plt.axvline(mean_Q3, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean: {mean_Q3:.2f}')
plt.axvline(median_Q3, color='yellow', linestyle='dashed', linewidth=1.5, label=f'Median: {median_Q3:.2f}')
plt.xlabel("Marks scored in Q3")
plt.ylabel("No. of students")
plt.title("Histogram of Q3 with Mean and Median")
plt.legend()
plt.tight_layout()
plt.show()
```

The red dashed line indicates the mean, while the yellow dashed line represents the median. 
If the mean is far from the median, it suggests the presence of skewness in the data.

```python
df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
df["Q3"] = df["Q3aM5"] + df["Q3bM5"]
df["Q4"] = df["Q4aM3"] + df["Q4bM7"]
df["Q5"] = df["Q5M10"]
df["Q6"] = df["Q6aM4"] + df["Q6bM6"]
columns_to_drop = ["Q1aM4", "Q1bM6", "Q2aM6", "Q2bM4", "Q3aM5", "Q3bM5", "Q4aM3", "Q4bM7", "Q5M10", "Q6aM4", "Q6bM6","Score_Category"]
df = df.drop(columns=columns_to_drop)
df
```

```python
import numpy as np
numeric_df = df.select_dtypes(include=['number'])\
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(ticks=np.arange(len(correlation_matrix.columns)), labels=correlation_matrix.columns, rotation=45)
plt.yticks(ticks=np.arange(len(correlation_matrix.columns)), labels=correlation_matrix.columns)
plt.title("Correlation Matrix")
plt.show()
```

Positive correlations (closer to +1) will appear in shades of red, while negative correlations 
(closer to -1) will appear in shades of blue.The diagonal elements will always be 1 (self correlation)
this visualization toidentify strongly correlated variables, which may influence analysis 
or modeling decision

```python
def performance_category(total):
    if total >= 35:
        return "Excellent"
    elif total >= 30:
        return "Good"
    elif total >= 20:
        return "Average"
    else:
        return "Poor"
df['Performance'] = df['Total'].apply(performance_category)
print(df[['Total', 'Performance']])
```
